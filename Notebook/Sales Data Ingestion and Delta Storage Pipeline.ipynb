```python
from pyspark.sql.functions import *
from pyspark.sql import *
```

```python
spark.conf.set(
  "fs.azure.account.key.onlineaccstorage.dfs.core.windows.net",
  dbutils.secrets.get(scope="onlinesales", key="storagevalut")
)
```

```python
storage_api = "abfss://rawdata@onlineaccstorage.dfs.core.windows.net/"
dbutils.fs.ls(storage_api)
```

```python
storage = "abfss://salesdata@onlineaccstorage.dfs.core.windows.net/"
bronze_path = storage + "bronze/"

accounts_store = bronze_path + "accounts/"
product_store = bronze_path + "product/"
data_dictionary = bronze_path + "data_dictionary/"
sales_pipeline_store = bronze_path + "sales_pipeline/"
sales_teams_store = bronze_path + "sales_teams/"
```

```python
accounts = spark.read.format("csv").option("header", True).load(storage_api + "dbo.accounts.csv")
products = spark.read.format("csv").option("header", "true").load(storage_api + "dbo.products.csv")
data_dictionary = spark.read.format("csv").option("header", True).load(storage_api + "dbo.data_dictionary.csv")
sales_pipeline = spark.read.format("csv").option("header", True).load(storage_api + "dbo.sales_pipeline.csv")
sales_teams = spark.read.format("csv").option("header", True).load(storage_api + "dbo.sales_teams.csv")
```

```python
data_dictionary.write.mode("overwrite").format("delta").option("overwriteSchema", True).option("mergeSchema", True).save(bronze_path + "data_dictionary/")
sales_pipeline.write.mode("overwrite").format("delta").option("overwriteSchema", True).option("mergeSchema", True).save(sales_pipeline_store)
sales_teams.write.mode("overwrite").format("delta").option("overwriteSchema", True).option("mergeSchema", True).save(sales_teams_store)
products.write.mode("overwrite").format("delta").option("overwriteSchema", True).option("mergeSchema", True).save(product_store)
accounts.write.mode("overwrite").format("delta").option("overwriteSchema", True).option("mergeSchema", True).save(accounts_store)
```
